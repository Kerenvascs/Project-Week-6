## Goodness of fit
Association test between each variable and response
- Test if data observed os uniformely distributed
 - If is uniform P(A) = P(B) = P(C) = P(n)

```python
df = bank[['marital', 'y']]
# df.groupby(['job', 'y']).count())
df.groupby(['marital','y'])['y'].count()
```

```python
observed = np.array(df.groupby(['marital','y'])['y'].count())
len(observed)
```

```python
# expected frequencies -> assuming that quantity sold is uniformly distributed
# expected = n . PTheory 1/24
expected = np.array([observed.sum()/6, observed.sum()/6,observed.sum()/6,observed.sum()/6,observed.sum()/6,observed.sum()/6] )
expected
```

```python
# perform chi-squared test 
stats.chisquare(observed, expected)
```
